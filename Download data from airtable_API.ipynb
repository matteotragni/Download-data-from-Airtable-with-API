```python
import requests
import os
import zipfile
import re

# Airtable API credentials
api_key = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX' # replace with your API key
base_id = 'appCIpKuHLVqt1oTl' # replace (you find in the URL)
table_id = 'tblKy625bnQiEgTg1' # replace (you find it in the URL)

# Headers for the Airtable API request
headers = {
    'Authorization': f'Bearer {api_key}'
}

# Airtable API URL
url = f'https://api.airtable.com/v0/{base_id}/{table_id}'

# Function to sanitize file names
def sanitize_filename(filename):
    # Remove any characters that are not alphanumeric, dots, underscores, or hyphens
    return re.sub(r'[^A-Za-z0-9._-]', '_', filename)

# Function to get the path to the Desktop
def get_desktop_path():
    home = os.path.expanduser("~")
    return os.path.join(home, "Desktop")

# Function to download files and save them in a zip file
def download_and_zip_files(zip_directory):
    # Make a request to the Airtable API to get all records
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()

    # Ensure the directory exists
    if not os.path.exists(zip_directory):
        os.makedirs(zip_directory)

    # Full path to the zip file
    zip_file_path = os.path.join(zip_directory, 'downloaded_files.zip')

    # Create a ZipFile object
    with zipfile.ZipFile(zip_file_path, 'w') as zipf:
        # Iterate through each record
        for record in data['records']:
            fields = record['fields']
            
            if 'Allegati' in fields and 'File name' in fields:
                attachments = fields['Allegati']
                file_name = sanitize_filename(fields['File name'])
                
                # Download each attachment
                for attachment in attachments:
                    file_url = attachment['url']
                    original_file_name = sanitize_filename(attachment['filename'])
                    
                    # Download the file
                    print(f'Downloading {original_file_name}...')
                    file_response = requests.get(file_url)
                    file_response.raise_for_status()

                    # Save the file with the new name in memory
                    new_file_name = f"{file_name}{os.path.splitext(original_file_name)[1]}"
                    file_path = os.path.join(zip_directory, new_file_name)
                    with open(file_path, 'wb') as file:
                        file.write(file_response.content)
                    
                    # Add the file to the zip
                    zipf.write(file_path, new_file_name)
                    
                    # Clean up the local file after adding it to the zip
                    os.remove(file_path)
                    
                    print(f'Saved and zipped as {new_file_name}')

    print(f'ZIP file created at: {zip_file_path}')

if __name__ == "__main__":
    # Get the path to the Desktop
    desktop_path = get_desktop_path()
    download_and_zip_files(desktop_path)
```
